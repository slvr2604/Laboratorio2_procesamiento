```python
#Se declaran las variables.
Ts = 1.25e-3
f0 = 100
n = np.arange(9)


x1 = np.cos(2*np.pi*f0*n*Ts)
x2 = np.sin(2*np.pi*f0*n*Ts)

# Calculamos la correlación cruzada.
N = len(x1)
r = []
lags = range(-N+1, N)

for k in lags:
    suma = 0
    for i in range(N):
        j = i + k
        if 0 <= j < N:
            suma += x1[i] * x2[j]
    r.append(suma)

print("Valores de la correlación cruzada:")
for k, val in zip(lags, r):
    print(f"k={k:2d}  =  {val:.4f}")

# Graficamos
plt.stem(lags, r)
plt.title("Correlación cruzada")
plt.xlabel("Retardo k")
plt.ylabel("r (x1x2)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos de la señal
ruta = "/content/drive/MyDrive/Colab Notebooks/senal.txt"
senal = np.loadtxt(ruta) * 1000  # paso a mV

# Defino frecuencia de muestreo (ajustar si no es 1000 Hz)
fs = 1000
tiempo = np.arange(len(senal)) / fs

# Calculo estadísticos básicos
print("=== DATOS ESTADÍSTICOS DE LA SEÑAL ===")
print("Número de datos:", len(senal))
print("Duración de la señal:", round(tiempo[-1], 3), "segundos")
print("Media:", round(np.mean(senal), 4), "mV")
print("Mediana:", round(np.median(senal), 4), "mV")
print("Desviación estándar:", round(np.std(senal), 4), "mV")
print("Valor mínimo:", round(np.min(senal), 4), "mV")
print("Valor máximo:", round(np.max(senal), 4), "mV")

# Convolución con una ventana simple (promedio)
ventana = np.ones(50)/50
senal_suavizada = np.convolve(senal, ventana, mode='same')

# Autocorrelación de la señal
auto_corr = np.correlate(senal - np.mean(senal), senal - np.mean(senal), mode='full')
auto_corr = auto_corr / np.max(auto_corr)
lags = np.arange(-len(senal)+1, len(senal)) / fs

# Graficar señal original
plt.figure(figsize=(12,4))
plt.plot(tiempo, senal)
plt.title("Señal EOG en el tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [mV]")
plt.grid()
plt.show()

# Graficar convolución
plt.figure(figsize=(12,4))
plt.plot(tiempo, senal_suavizada, color="orange")
plt.title("Señal EOG suavizada (convolución)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [mV]")
plt.grid()
plt.show()

# Graficar correlación
plt.figure(figsize=(12,4))
plt.plot(lags, auto_corr, color="purple")
plt.title("Autocorrelación de la señal")
plt.xlabel("Desfase [s]")
plt.ylabel("Correlación normalizada")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# === 4a. Transformada de Fourier ===
fft = np.fft.fft(senal - np.mean(senal))
freqs = np.fft.fftfreq(len(fft), 1/1200)
fft_magnitud = np.abs(fft)

# Tomo solo frecuencias positivas hasta 100 Hz
mask = (freqs >= 0) & (freqs <= 100)
f_pos = freqs[mask]
mag_pos = fft_magnitud[mask]

plt.figure(figsize=(12,4))
plt.plot(f_pos, mag_pos)
plt.title("Transformada de Fourier (0-100 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")
plt.grid()
plt.show()

# === 4b. Densidad Espectral de Potencia (PSD) ===
f_psd, Pxx = welch(senal, 1200, nperseg=1024)

plt.figure(figsize=(12,4))
plt.semilogy(f_psd[f_psd <= 100], Pxx[f_psd <= 100])
plt.title("Densidad Espectral de Potencia (0-100 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Potencia [V²/Hz]")
plt.grid()
plt.show()

# === 4c. Estadísticos en Frecuencia ===
freq_media = np.sum(f_pos * mag_pos) / np.sum(mag_pos)
acumulada = np.cumsum(mag_pos)
freq_mediana = f_pos[np.where(acumulada >= acumulada[-1]/2)[0][0]]
freq_std = np.sqrt(np.sum(((f_pos - freq_media)**2) * mag_pos) / np.sum(mag_pos))

print("=== Estadísticos en Frecuencia (0-100 Hz) ===")
print("Frecuencia media:", round(freq_media, 2), "Hz")
print("Frecuencia mediana:", round(freq_mediana, 2), "Hz")
print("Desviación estándar:", round(freq_std, 2), "Hz")

# Histograma de frecuencias
plt.figure(figsize=(12,4))
plt.hist(f_pos, bins=50, weights=mag_pos)
plt.xlim(0, 80)
plt.title("Histograma de Frecuencias (0-100 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Energía relativa")
plt.grid()
plt.show()
```
