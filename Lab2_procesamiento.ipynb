```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos x(n) y h(n).
x = np.array([1,0,5,4,2,8,4,0,8,8])
h = np.array([5,6,0,0,8,5,6])

y = np.convolve(x,h)
print("y(n) = ", y)
```

```python
#Generamos el rango de índices.
x_n = np.arange(len(x))
h_n = np.arange(len(h))
y_n = np.arange(len(y))

print ("Los índices de la convolución son: ", x_n, h_n, y_n)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=False)

# Gráfica de x(n)
plt.subplot(3, 1, 1)
plt.stem(x_n, x)
plt.title("x(n) = Cédula Silvia")
plt.ylabel("x[n]")
plt.grid(True)

# Gráfica de h(n)
plt.subplot(3, 1, 2)
plt.stem(h_n, h)
plt.title("h(n) = Código Silvia")
plt.ylabel("h[n]")
plt.grid(True)

# Gráfica de y(n)
plt.subplot(3, 1, 3)
plt.stem(y_n, y)
plt.title("y(n) = Convolución x(n) + y(n)")
plt.xlabel("n")
plt.ylabel("y[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Se declaran las variables.
Ts = 1.25e-3
f0 = 100
n = np.arange(9)


x1 = np.cos(2*np.pi*f0*n*Ts)
x2 = np.sin(2*np.pi*f0*n*Ts)

# Calculamos la correlación cruzada.
N = len(x1)
r = []
lags = range(-N+1, N)

for k in lags:
    suma = 0
    for i in range(N):
        j = i + k
        if 0 <= j < N:
            suma += x1[i] * x2[j]
    r.append(suma)

print("Valores de la correlación cruzada:")
for k, val in zip(lags, r):
    print(f"k={k:2d}  =  {val:.4f}")

# Graficamos
plt.stem(lags, r)
plt.title("Correlación cruzada")
plt.xlabel("Retardo k")
plt.ylabel("r (x1x2)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos de la señal
ruta = "/content/drive/MyDrive/Colab Notebooks/senal.txt"
senal = np.loadtxt(ruta) * 1000  # paso a mV

# Defino frecuencia de muestreo (ajustar si no es 1000 Hz)
fs = 1000
tiempo = np.arange(len(senal)) / fs


# Convolución con una ventana simple (promedio)
ventana = np.ones(50)/50
senal_suavizada = np.convolve(senal, ventana, mode='same')

# Autocorrelación de la señal
auto_corr = np.correlate(senal - np.mean(senal), senal - np.mean(senal), mode='full')
auto_corr = auto_corr / np.max(auto_corr)
lags = np.arange(-len(senal)+1, len(senal)) / fs

# Graficar convolución
plt.figure(figsize=(12,4))
plt.plot(tiempo, senal_suavizada, color="orange")
plt.title("Señal EOG suavizada (convolución)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [mV]")
plt.grid()
plt.show()

# Graficar correlación
plt.figure(figsize=(12,4))
plt.plot(lags, auto_corr, color="purple")
plt.title("Autocorrelación de la señal")
plt.xlabel("Desfase [s]")
plt.ylabel("Correlación normalizada")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# === 1. Cargar la señal ===
ruta_archivo = "/content/drive/MyDrive/Colab Notebooks/senal.txt"
senal_mV = np.loadtxt(ruta_archivo) * 1000  # señal en mV
frecuencia_muestreo = 1200  # Hz

# === 2. Transformada de Fourier ===
fft_resultado = np.fft.fft(senal_mV - np.mean(senal_mV))
frecuencias_fft = np.fft.fftfreq(len(fft_resultado), 1/frecuencia_muestreo)
magnitud_fft = np.abs(fft_resultado)

# Filtrar solo frecuencias de 0 a 100 Hz
filtro = (frecuencias_fft >= 0) & (frecuencias_fft <= 100)
frecuencias_filtradas = frecuencias_fft[filtro]
magnitud_filtrada = magnitud_fft[filtro]

plt.figure(figsize=(12,4))
plt.plot(frecuencias_filtradas, magnitud_filtrada)
plt.title("Transformada de Fourier (0-100 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")
plt.grid()
plt.show()

# === 3. Densidad Espectral de Potencia (PSD) ===
frecuencias_psd, potencia_psd = welch(senal_mV, frecuencia_muestreo, nperseg=1024)

plt.figure(figsize=(12,4))
plt.semilogy(frecuencias_psd[frecuencias_psd <= 100], potencia_psd[frecuencias_psd <= 100])
plt.title("Densidad Espectral de Potencia (0-100 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Potencia [mV²/Hz]")
plt.grid()
plt.show()

# === 4. Estadísticos en Frecuencia ===
frecuencia_media = np.sum(frecuencias_filtradas * magnitud_filtrada) / np.sum(magnitud_filtrada)
magnitud_acumulada = np.cumsum(magnitud_filtrada)
frecuencia_mediana = frecuencias_filtradas[np.where(magnitud_acumulada >= magnitud_acumulada[-1]/2)[0][0]]
desviacion_frecuencia = np.sqrt(np.sum(((frecuencias_filtradas - frecuencia_media)**2) * magnitud_filtrada) / np.sum(magnitud_filtrada))

print("=== Estadísticos en Frecuencia (0-100 Hz) ===")
print(f"Frecuencia media: {frecuencia_media:.2f} Hz")
print(f"Frecuencia mediana: {frecuencia_mediana:.2f} Hz")
print(f"Desviación estándar: {desviacion_frecuencia:.2f} Hz")

# === 5. Histograma de Frecuencias ===
plt.figure(figsize=(12,4))
plt.hist(frecuencias_filtradas, bins=50, weights=magnitud_filtrada)
plt.xlim(0, 80)
plt.title("Histograma de Frecuencias (0-80 Hz)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Energía relativa")
plt.grid()
plt.show()
```
